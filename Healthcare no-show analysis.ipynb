## Data Cleaning

```python
# Standardize column names 
df.columns = (
    df.columns
        .str.strip()
        .str.replace('-', '_', regex=False)
        .str.replace(' ', '_', regex=False)
        .str.lower()
)
```

```python
df.columns
```

```python
# Fix issues
df.rename(columns={
    "patientid" : "patient_id",
    "appointmentid" : "appointment_id",
    "hipertension": "hypertension",
    "handcap": "handicap",
    "scheduledday":"scheduled_day",
    "appointmentday" : "appointment_day"
    
}, inplace=True)
```

```python
df.columns
```

```python
# Convert date columns
df["scheduled_day"] = pd.to_datetime(df["scheduled_day"])
df["appointment_day"] = pd.to_datetime(df["appointment_day"])

```

```python
# create waiting_days feature
df["waiting_days"] = (df["appointment_day"] - df["scheduled_day"]).dt.days
```

```python
#  Remove negative waiting days if any
df = df[df["waiting_days"] >= 0].copy()
```

```python
# Convert No_show into binary
df['No_show_flag'] = df['no_show'].map({'Yes': 1, 'No': 0})
```

 # Feature Engineering

```python
# Age Group
bins = [-1, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# Waiting bucket
df['waiting_bucket'] = pd.cut(df['waiting_days'],
                            bins = [-1, 1, 7, 30, float('inf')],
                             labels=['Same Day', '1-7 Days', '1-4 Weeks', '1+ Month'])
```

# EDA

```python
# Overall No-show Rate

total_appointments = len(df)
print("Total Appointments:", total_appointments)

no_show_rate = df['No_show_flag'].mean() * 100
print("No-Show Rate (%):", round(no_show_rate, 2))

plt.figure(figsize=(5,5))
plt.pie([no_show_rate, 100-no_show_rate],
        labels=['No-Show', 'Show'],
        autopct='%1.1f%%')
plt.title("Overall No-Show Rate")
plt.show()
```

```python
# Waiting Days Impact

waiting_analysis = df.groupby("waiting_bucket", observed=True)['No_show_flag'].mean() * 100


print(waiting_analysis.round(2))

plt.figure(figsize=(8,5))
waiting_analysis.plot(kind='bar')
plt.title("No-Show Rate by Waiting Time")
plt.xlabel("Waiting Bucket")
plt.ylabel("No-Show %")
plt.xticks(rotation=0)
plt.show()
```

```python
# Age Group Impact

age_analysis = df.groupby("age_group", observed=True)['No_show_flag'].mean() * 100

print("\n===== AGE GROUP IMPACT =====")
print(age_analysis.round(2))

plt.figure(figsize=(8,5))
age_analysis.plot(kind='bar', color='blue')
plt.title("No-Show Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("No-Show %")
plt.xticks(rotation=0)
plt.show()
```

```python
# SMS Impact

sms_analysis = df.groupby("sms_received")['No_show_flag'].mean() * 100


print(sms_analysis.round(2))

plt.figure(figsize=(6,4))
sms_analysis.plot(kind='bar', color='green')
plt.title("Impact of SMS on No-Show")
plt.xlabel("SMS Received ")
plt.ylabel("No-Show %")
plt.xticks(rotation=0)
plt.show()
```

```python

# Waiting Days Distribution

plt.figure(figsize=(8,5))
sns.histplot(df['waiting_days'], bins=30)
plt.title("Distribution of Waiting Days")
plt.xlabel("Waiting Days")
plt.ylabel("Count")
plt.show()
```

```python
# Assume average revenue per patient visit is ₹1500
avg_revenue_per_patient = 1500

# Calculate total number no-shows
total_no_shows = df['No_show_flag'].sum()
print("Total No-Shows:", total_no_shows)

# Estimate total revenue loss due to missed appointments

estimated_loss = total_no_shows * avg_revenue_per_patient
print("Estimated Revenue Loss (₹):", estimated_loss)
```

# Insight And Recommendation

## Insight
1. Long waiting time is the primary driver of patient no-shows. 

2. Patients with higher waiting days are significantly more likely to miss their appointments,leading to operational inefficiencies and revenue loss.

## Recommendation 
1. Hospitals should focus on reducing waiting time through better scheduling and capacity planning. 

2. Additionally, implementing automated SMS reminders can further reduce no-shows and improve patient attendance.
